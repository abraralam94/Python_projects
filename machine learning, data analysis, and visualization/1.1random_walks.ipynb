```python
from random import choice

class RandomWalk:
    ##A class to generate random walks.
    
    def __init__(self, num_points=5000):
        ##Initialize attributes of a walk."""
        self.num_points = num_points
        
        # All walks start at (0, 0).
        self.x_values = [0]
        self.y_values = [0]

    def fill_walk(self):
        ##Calculation of all the points in the walk."""
    
        # Keep taking steps until the walk reaches the desired length.
        while len(self.x_values) < self.num_points:
        
            # Decide which direction to go and how far to go in that direction.
            x_direction = choice([1, -1])
            x_distance = choice([0, 1, 2, 3, 4])
            x_step = x_direction * x_distance
        
            y_direction = choice([1, -1])
            y_distance = choice([0, 1, 2, 3, 4])
            y_step = y_direction * y_distance
        
            # Reject moves that go nowhere.
            if x_step == 0 and y_step == 0:
                continue
        
            # Calculate the new position.
            x = self.x_values[-1] + x_step
            y = self.y_values[-1] + y_step
        
            self.x_values.append(x)
            self.y_values.append(y)
```

The code below is for visualization of the random walk's random behaviour

```python
import matplotlib.pyplot as plt


while True:
	rw = RandomWalk(500)
	rw.fill_walk()

	# Plot the points in the walk
	plt.style.use('seaborn')
	fig, ax = plt.subplots()
	point_numbers = range (rw.num_points)
	ax.scatter(rw.y_values, rw.x_values,c = point_numbers, cmap = plt.cm.Blues, edgecolors = 'none', s = 30)
	ax.plot(rw.y_values, rw.x_values, linewidth = 3)## Models the movement of a pollen particle over water
	# Emphasize the first and last points
	ax.scatter(0, 0, c = 'green', edgecolors = 'none', s = 150)
	ax.scatter(rw.x_values[-1], rw.y_values[-1], c = 'red', edgecolors = 'none', s = 150)
	#removal of Axes
	ax.get_xaxis().set_visible(False)
	ax.get_yaxis().set_visible(False)
	#Now lets display the plot
	plt.show()

	#Prompt user
	response = input("Do you want to generate another random walk ? ('y'/'n')")
	if (response == 'n'):
		break
```
