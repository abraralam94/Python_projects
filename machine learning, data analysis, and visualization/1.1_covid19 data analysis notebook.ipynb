### task 3.2: find maxmimum infection rate for China

### Task 3.3: find maximum infection rate for all of the countries. 

```python
countries = list(corona_dataset_aggregated.index)
max_infection_rates = []
for country in countries:
    max = corona_dataset_aggregated.loc[country].diff().max()
    max_infection_rates.append(max)
max_infection_rates
```

```python
corona_dataset_aggregated["Max infection rate"] = max_infection_rates
corona_dataset_aggregated
```

### Task 3.4: create a new dataframe with only needed column 

```python
# We remove the specific month and date column entries
corona_data = pd.DataFrame(corona_dataset_aggregated["Max infection rate"])
```

### Task 5.1: Plotting max infection rate vs GDP per capita

```python
x = data["GDP per capita"]
```

```python
y = data["Max infection rate"]
```

```python
sns.scatterplot(x, y)
```

##### The plot above does not give any clear picture. However,before dismissing any correlation, we opt for taking the log of Max infection rate, since our scaling is drastically different for x and y axis


```python
sns.scatterplot(x, np.log(y))
```

```python
sns.regplot(x, np.log(y))
```

##### The plot above now depicts a clearer picture of a positive correlation. This does not necessarily mean however that one causes the other
